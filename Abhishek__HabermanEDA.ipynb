# UNIVARIATE ANALYSIS

# Making CDF to understand the data better

```python
#PLotting CDF based on 'Age_Of_Patient'

counts, bin_edges = np.histogram(cancer_survival['Age_Of_Patient'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)

#compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)

plt.show();
```

# Observation

* From the plot we can infer what percentage of people lie below what age

```python
#Plotting CDF based on 'Year_Of_Operation'

counts, bin_edges = np.histogram(cancer_survival['Year_Of_Operation'], bins=15, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)

#compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)

plt.show();
```

# Observation

* From the plot we can infer what percentage of people were operated before which year

```python
#Plotting CDF based on 'Positive Axilliary Nodes'

counts, bin_edges = np.histogram(cancer_survival['Pos_Ax_Nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)

#compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)

plt.show();
```

# Observation

* From the plot we can clearly infer that 80% of the people have 0 Positive Axilliary nodes.
